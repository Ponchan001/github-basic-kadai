# 新しいセクション

```python
import pandas as pd

# Excelファイルを読み込む
# 例: 入力ファイル名が「売上.xlsx」の場合
df = pd.read_excel("売上.xlsx")

# 売上列の平均値を算出
average_sales = df["売上"].mean()

# 平均売上列を追加
df["平均売上"] = average_sales

# 業績ランクを判定する関数
def performance(sales):
    if sales >= average_sales + 50:
        return "A"
    elif sales >= average_sales:
        return "B"
    else:
        return "C"

# 業績ランク列を追加
df["業績ランク"] = df["売上"].apply(performance)

# 新しいExcelファイルとして保存
output_file = "業績.xlsx"
df.to_excel(output_file, index=False)

print(f"出力完了: {output_file}")
```

```python
import pandas as pd

def main(file_path: str):
    df = pd.read_excel(file_path)

    average_sales = df["売上"].mean()
    df["平均売上"] = average_sales

    def performance(sales):
        if sales >= average_sales + 50:
            return "A"
        elif sales >= average_sales:
            return "B"
        else:
            return "C"

    df["業績ランク"] = df["売上"].apply(performance)

    output_path = "業績.xlsx"
    df.to_excel(output_path, index=False)

    return {
        "平均売上": average_sales,
        "出力ファイル": output_path
    }
```
