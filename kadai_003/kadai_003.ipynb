```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from google.colab import files

# ==========================
# Excelファイルをアップロード
# ==========================
uploaded = files.upload()

# ==========================
# Excelファイルの読み込み
# ==========================
df2022 = pd.read_excel("2022_sales.xlsx", sheet_name="Sheet1")
df2023 = pd.read_excel("2023_sales.xlsx", sheet_name="Sheet1")

# ==========================
# 列名を統一
# ==========================
def normalize_columns(df):
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace("(", "（", regex=False)
        .str.replace(")", "）", regex=False)
    )
    return df

df2022 = normalize_columns(df2022)
df2023 = normalize_columns(df2023)

# ==========================
# 必要な列があるか確認
# ==========================
required_columns = ["商品", "売上年", "金額（千円）"]

for col in required_columns:
    if col not in df2022.columns:
        raise ValueError(f"2022_sales.xlsx に '{col}' 列がありません。")

for col in required_columns:
    if col not in df2023.columns:
        raise ValueError(f"2023_sales.xlsx に '{col}' 列がありません。")

# ==========================
# データ連結
# ==========================
df = pd.concat([df2022, df2023], ignore_index=True)

# ==========================
# 商品・売上年ごとに合計
# ==========================
summary = (
    df.groupby(["商品", "売上年"], as_index=False)["金額（千円）"]
      .sum()
)

# ==========================
# Excelへ保存
# ==========================
output_file = "売上集計表.xlsx"

summary.to_excel(output_file, index=False)

# ==========================
# ヘッダーを薄いグレー(#F2F2F2)に設定
# ==========================
wb = load_workbook(output_file)
ws = wb.active

fill = PatternFill(
    fill_type="solid",
    start_color="F2F2F2",
    end_color="F2F2F2"
)

for cell in ws[1]:
    cell.fill = fill

wb.save(output_file)

print(f"{output_file} を作成しました。")

# ==========================
# ダウンロード
# ==========================
files.download(output_file)
```
