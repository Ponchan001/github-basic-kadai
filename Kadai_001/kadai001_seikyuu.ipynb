```python
from openpyxl import Workbook
from openpyxl.styles import Font, Border, Side, Alignment
from openpyxl.worksheet.page import PageMargins

# ブック作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# -----------------------
# 列幅
# -----------------------
widths = {
    "A": 5,
    "B": 18,
    "C": 8,
    "D": 12,
    "E": 12,
    "F": 8,
    "G": 15
}

for col, w in widths.items():
    ws.column_dimensions[col].width = w

# 行高さ
for i in range(1, 18):
    ws.row_dimensions[i].height = 22

# -----------------------
# フォント
# -----------------------
title_font = Font(size=18, bold=True)
bold_font = Font(size=12, bold=True)
normal_font = Font(size=12)

# -----------------------
# タイトル
# -----------------------
ws["B2"] = "請求書"
ws["B2"].font = title_font

# -----------------------
# 会社情報
# -----------------------
ws["B4"] = "株式会社ABC"
ws["B4"].font = bold_font

ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# -----------------------
# No・日付
# -----------------------
ws["F4"] = "No."
ws["F4"].font = bold_font

ws["G4"] = "0001"
ws["G4"].font = bold_font

ws["F5"] = "日付"
ws["F5"].font = bold_font

ws["G5"] = "2023/06/06"

# -----------------------
# 商品表
# -----------------------
headers = ["商品名", "数量", "単価", "金額"]

for col, value in zip(["B","C","D","E"], headers):
    cell = ws[f"{col}10"]
    cell.value = value
    cell.font = bold_font
    cell.alignment = Alignment(horizontal="center")

# データ
data = [
    ["商品A",2,10000,20000],
    ["商品B",1,15000,15000]
]

row = 11

for item in data:

    ws[f"B{row}"] = item[0]

    ws[f"C{row}"] = item[1]

    ws[f"D{row}"] = item[2]

    ws[f"E{row}"] = item[3]

    row += 1

# 合計行
ws["E13"] = "=SUM(E11:E12)"

ws["B15"] = "小計"
ws["E15"] = "=E13"

ws["B16"] = "消費税"
ws["E16"] = "=ROUND(E15*0.1,0)"

ws["B17"] = "合計"
ws["E17"] = "=E15+E16"

# -----------------------
# 数値書式
# -----------------------
for r in range(11,18):
    for c in ["C","D","E"]:
        ws[f"{c}{r}"].number_format = '#,##0'
        ws[f"{c}{r}"].alignment = Alignment(horizontal="right")

# -----------------------
# 罫線
# -----------------------
thin = Side(style="thin")

border = Border(
    left=thin,
    right=thin,
    top=thin,
    bottom=thin
)

for row in range(10,17):
    for col in range(2,6):
        ws.cell(row=row,column=col).border = border

# -----------------------
# 印刷設定
# -----------------------
ws.page_setup.paperSize = ws.PAPERSIZE_A4
ws.page_setup.orientation = "portrait"

ws.page_margins = PageMargins(
    left=0.3,
    right=0.3,
    top=0.5,
    bottom=0.5
)

# -----------------------
# 保存
# -----------------------
wb.save("請求書完成版.xlsx")

print("請求書完成版.xlsx を作成しました。")
```
